```python
%matplotlib inline

# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set Filepath for both CSV Files
filepath1 = "../data/mouse_drug_data.csv"
filepath2 = "../data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(filepath1)

clinical_trial_data = pd.read_csv(filepath2)
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(clinical_trial_data, mouse_drug_data, on="Mouse ID", how="left")

# Display the data table for preview
combined_data.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
average_tumor_volume = combined_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
average_tumor_volume = average_tumor_volume.reset_index()

# Preview DataFrame
average_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_standard_errors = combined_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_volume_standard_errors = tumor_volume_standard_errors.reset_index()

# Preview DataFrame
tumor_volume_standard_errors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
average_tumor_volume2 = average_tumor_volume.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Preview that Reformatting worked
average_tumor_volume2.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_volume_standard_errors2 = tumor_volume_standard_errors.pivot(index="Timepoint", columns="Drug"
                                                                 , values="Tumor Volume (mm3)")

# Preview that Reformatting worked
tumor_volume_standard_errors2.head(10)
```

```python
# Generate the Plot (with Error Bars)
drug_list = [("Capomulin", "o", "red"), ("Infubinol", "^", "blue"), ("Ketapril", "s", "green"), ("Placebo", "d", "black")]

for drug, marker, color in drug_list:
    
    tumor_treatment_graph = plt.errorbar(average_tumor_volume2.index, average_tumor_volume2[drug]
                                      , tumor_volume_standard_errors2[drug] , marker=marker, color=color
                                      , ls="--", linewidth=0.5, label=drug)

plt.plot()
plt.grid()
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="upper left")

# Save the Figure
plt.savefig("../Output_Images/tumor_response_to_treatment.png")

# Show the Figure
plt.show()
plt.close()
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
average_metastatic_sites = combined_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
average_metastatic_sites = average_metastatic_sites.reset_index()

# Preview DataFrame
average_metastatic_sites.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sites_standard_errors = combined_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
metastatic_sites_standard_errors = metastatic_sites_standard_errors.reset_index()

# Preview DataFrame
metastatic_sites_standard_errors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
average_metastatic_sites2 = average_metastatic_sites.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
average_metastatic_sites2.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_sites_standard_errors2 = metastatic_sites_standard_errors.pivot(index="Timepoint", columns="Drug"
                                                                         , values="Metastatic Sites")

# Preview that Reformatting worked
metastatic_sites_standard_errors2.head(10)
```

```python
# Generate the Plot (with Error Bars)
drug_list = [("Capomulin", "o", "red"), ("Infubinol", "^", "blue"), ("Ketapril", "s", "green"), ("Placebo", "d", "black")]

for drug, marker, color in drug_list:

    metastatic_spread_graph = plt.errorbar(average_metastatic_sites2.index, average_metastatic_sites2[drug]
                                         , metastatic_sites_standard_errors2[drug], marker=marker
                                         , color=color, ls="--", linewidth=0.5, label=drug)

plt.plot()
plt.grid()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.legend(loc="upper left")

# Save the Figure
plt.savefig("../Output_Images/metastatic_spread_during_treatment.png")


# Show the Figure
plt.show()
plt.close()
```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = combined_data.groupby(["Drug", "Timepoint"])["Mouse ID"].nunique()

# Convert to DataFrame
mice_count = mice_count.reset_index()

mice_count = mice_count.rename(columns={"Mouse ID": "Mouse Count"})

# Preview DataFrame
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count2 = mice_count.pivot(index="Timepoint", columns="Drug", values="Mouse Count")

# Preview the Data Frame
mice_count2.head(10)
```

```python
# Generate the Plot (Accounting for percentages)
drug_list = [("Capomulin", "o", "red"), ("Infubinol", "^", "blue"), ("Ketapril", "s", "green"), ("Placebo", "d", "black")]

for drug, marker, color in drug_list:
    
    total_mouse_count = mice_count2[drug][0]
    
    survival_rate = (mice_count2[drug] / total_mouse_count) * 100
    
    mouse_survival_rate_graph = plt.plot(mice_count2.index, survival_rate, marker=marker, color=color
                                   , ls="--", linewidth=0.5, label=drug)

plt.plot()
plt.grid()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc="lower left")

# Save the Figure
plt.savefig("../Output_Images/survival_during_treatment.png")

# Show the Figure
plt.show()
plt.close()
```

```python
# Calculate the percent changes for each drug
percent_change = ((average_tumor_volume2.iloc[-1, :] - average_tumor_volume2.iloc[0, :]) / average_tumor_volume2.iloc[0, :]) * 100

# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change_tuple = tuple(percent_change)

# Splice the data between passing and failing drugs
passing_drugs = percent_change < 0

# Orient widths. Add labels, tick marks, etc. 
drug_list = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

percent_change_list = [(percent_change[drug]) for drug in drug_list]

percent_change_graph = plt.bar(drug_list, percent_change_list, color=passing_drugs.map({True: "green", False: "red"})
                             , width=-1, align="edge")

plt.grid()
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.ylim(-30, 70)

# Use functions to label the percentages of changes
def autolabel(rects):
    
    for rect in rects:
        
        height = rect.get_height()
        
        if height > 0:
            
            label_position = 4
            
        else:
            
            label_position = -8

        plt.text(rect.get_x() + rect.get_width()/2, label_position, f"{int(height)}%" , color="white", ha="center", va="bottom")
        
# Call functions to implement the function calls
autolabel(percent_change_graph)

# Save the Figure
plt.savefig("../Output_Images/tumor_change_over_45_day_treatment.png")


# Show the Figure
plt.show()
plt.close()
```
